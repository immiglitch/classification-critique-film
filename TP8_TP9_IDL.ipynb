```python
X_wine, y_wine = wine.data, wine.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_wine, y_wine, test_size=0.3)
y_train
```

```python
import matplotlib.pyplot as plt

plt.hist(y_train, align="right", label="train")
plt.hist(y_test, align="left", label="test")
plt.legend()
plt.xlabel("Classe")
plt.ylabel("Nombre d'exemples")
plt.title("Répartition des classes")
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_wine, y_wine, test_size=0.25, stratify=y_wine)
plt.hist(y_train, align="right", label="train")
plt.hist(y_test, align="left", label="test")
plt.legend()
plt.xlabel("Classe")
plt.ylabel("Nombre d'exemples")
plt.title("Répartition des classes avec échantillonnage stratifié")
plt.show()
```

```python
#Entraînement
from sklearn.svm import LinearSVC
clf = LinearSVC(dual=True)
clf.fit(X_train, y_train)

print(clf.predict(X_test))
```

```python
#Evaluation

print(clf.score(X_test, y_test))
from sklearn.metrics import classification_report

y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))

```

```python
# Exercice : refaire une partition train/test différente et comparer les résultats

from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, accuracy_score

# Liste des tailles de test à comparer
tailles_test = [0.20, 0.25, 0.30]

# On va stocker les résultats pour les comparer ensuite
resultats_partitions = []

for taille_test in tailles_test:

    # Nouvelle séparation train/test
    X_train, X_test, y_train, y_test = train_test_split(
        X_wine,
        y_wine,
        test_size=taille_test,
        random_state=42,
        stratify=y_wine
    )

    # Création du modèle
    clf = LinearSVC(dual=True, max_iter=10000)

    # Entraînement du modèle
    clf.fit(X_train, y_train)

    # Prédiction sur les données de test
    y_pred = clf.predict(X_test)

    # Calcul du score
    accuracy = accuracy_score(y_test, y_pred)

    # Sauvegarde du résultat
    resultats_partitions.append((taille_test, accuracy))

    print("--------------------------")
    print(f"Partition avec test_size = {taille_test}")
    print(f"Accuracy : {accuracy:.4f}")
    print(classification_report(y_test, y_pred))
```

```python
print("Résumé des résultats :")

for taille_test, accuracy in resultats_partitions:
    print(f"test_size = {taille_test} -> accuracy = {accuracy:.4f}")
```

### Interprétation

On observe que les scores peuvent légèrement varier selon la partition train/test utilisée.

Lorsque `test_size = 0.20`, le modèle dispose de plus de données pour l'entraînement, mais l'évaluation est faite sur moins d'exemples.

Lorsque `test_size = 0.30`, le modèle dispose de moins de données pour apprendre, mais le test est réalisé sur davantage d'exemples.

L'utilisation de `stratify=y_wine` permet de conserver une répartition équilibrée des classes dans les ensembles d'entraînement et de test. Cela rend la comparaison plus fiable.

Globalement, si les scores restent proches, cela signifie que le modèle est plutôt stable et ne dépend pas trop d'un seul découpage des données.
